### Task 1: A list of dictionaries 


```python
#data with transactions for Task 1
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

a) Filter out invalid transactions with price 0 or a negative amount of transactions

```python
#creating a new dictionary called validtransactions by filtering transactions using a for loop and if statement
validtransactions = [
    t for t in transactions 
        if t["price"] > 0 and t["quantity"] >= 0
]
        
```

```python
pprint(validtransactions)
```

b) Creating the field quantity

```python
#adding a new field with a for loop 
for t in validtransactions:
    t["total"] = t["price"] * t["quantity"]
```

```python
pprint(validtransactions)
```

c) Group transaction by region and calculate the total revenue per region

```python
regional_totals = {} #here the totals from the regions will be stored

for t in validtransactions: 
    region = t["region"]
    total = t["total"]

    if region in regional_totals:        # adding the totals from the regions
            regional_totals[region]+= total 
    else: 
        regional_totals[region] = total 
```

```python
pprint(regional_totals)
```

The transactions are grouped based on region by using a for loop and an if statement. The if statement checks if there exists a list with that region in the dictionary regional_total and adds the total from the region to the already existing ones. If there does not exist any, the if statement adds the new region to the dictionary. 


d) Printing the region with the highest total revenue 


```python
max_total = max(regional_totals.values())

for reg in regional_totals: 
    if regional_totals[reg] == max_total: 
        print(f"The region with the highest total revenue is {reg} with a total revenue of {max_total}") 
```

Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(sales):
    total_sales = sum(sales)
    average_sales = sum(sales)/len(sales)
    best_day = sales.index(max(sales)) + 1 #adding the + 1 to make results more logical since python starts indexing from 0 
    worst_day = sales.index(min(sales)) + 1 #adding the + 1 to make results more logical since python starts indexing from 0 
    above_average = sum(1 for sale in sales if sale > average_sales) #counting the sales that are fulfilling the condition and then assigning them with 1 

    #storing the values from the function in a dictonary
    return {
        "total": total_sales,
        "average": average_sales,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average
    } 
```

```python
summary = summarise_sales(sales)
pprint(summary)
```

### Task 5: Importing a module 

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

```python
from datetime import datetime

#converting the list into a date object 
dates = [datetime.strptime(order[0], "%Y-%m-%d").date() for order in orders]

#finding the first and last date and the calculating the difference
earliest = min(dates)
latest = max(dates)

difference = latest - earliest

print(f"Earliest order: {earliest}")
print(f"Latest order: {latest}")
print(f"Days between: {difference.days} days")
```

### Task 8: More Pandas 

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

a) Create DataFrame

```python
df_traffic = pd.DataFrame({
    "date": days, 
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df_traffic = df_traffic.set_index("date")

print(df_traffic)
```

b) Compute two new columns

```python
#computing conversions and creating the column, using round to round the value to nearest integer
df_traffic["conversions"] = (df_traffic["visitors"] * df_traffic["conversion_rate"]).round()

#computing daily revenue and creating the coulumn
df_traffic["daily_revenue"] = (df_traffic["conversions"] * df_traffic["avg_order_value"])

print(df_traffic[["conversions", "daily_revenue"]])
```
